```python
df = x.df.withColumn(
"sum_posteriorProbability",
f.expr("aggregate(locus, 0D, (acc, x) -> acc + x.posteriorProbability)")
)
```

```python
df.select("sum_posteriorProbability").describe().show()
```

```python
+-------+------------------------+                                              
|summary|sum_posteriorProbability|
+-------+------------------------+
|  count|                 2142311|
|   mean|      0.9781644863704999|
| stddev|    0.030583548286673787|
|    min|      0.9500000026761133|
|    max|      3.8397088152260723|
+-------+------------------------+
```

```python
df = df.withColumn(
    "locus",
    f.expr(
        """
        transform(locus, x -> struct(
            x.logBF as logBF,
            x.posteriorProbability / (sum_posteriorProbability/0.95000001) as posteriorProbability,
            x.variantId as variantId,
            x.pValueMantissa as pValueMantissa,
            x.pValueExponent as pValueExponent,
            x.beta as beta,
            x.standardError as standardError,
            x.r2Overall as r2Overall
        ))
        """
    )
).drop("sum_posteriorProbability")
```

```python
df = df.withColumn(
    "sum_posteriorProbability",
    f.expr("aggregate(locus, 0D, (acc, x) -> acc + x.posteriorProbability)")
)
```
