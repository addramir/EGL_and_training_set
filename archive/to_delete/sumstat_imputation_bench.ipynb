```python
import gentropy.method.sumstat_imputation as sumstat_imputation
```

```python
import numpy as np
import pandas as pd
ld=np.load("../../carmapy_light/tests/02_test_ld.npy")
ld=np.array(ld)
z=pd.read_csv("../../carmapy_light/tests/02_test_z.csv")
z=np.array(z.iloc[:,1])
```

```python
print(ld.shape)
print(len(z))
```

```python
import random

# Create a list of indices
indices = list(range(len(z)))

# Calculate 10% of the total indices
ten_percent = int(len(z) * 0.1)

# Randomly select 10% of the indices
unknowns = random.sample(indices, ten_percent)

# Get the remaining 90% of the indices
known = [index for index in indices if index not in unknowns]
```

```python
print(len(unknowns))
print(len(known))
```

```python
sig_t = ld[known,:][:,known]
sig_i_t = ld[unknowns, :][:,known]
zt = z[known]
```

```python
imp = sumstat_imputation.SummaryStatisticsImputation.raiss_model(zt, sig_t, sig_i_t, lamb=0.01, rtol=0.01)
```

```python
ind=imp["imputation_r2"]>0.95
print(sum(ind))
print(np.corrcoef(z[unknowns][ind],imp["mu"][ind]))

ind=imp["imputation_r2"]>0.8
print(sum(ind))
print(np.corrcoef(z[unknowns][ind],imp["mu"][ind]))
```
